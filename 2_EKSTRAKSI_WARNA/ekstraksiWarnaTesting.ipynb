```python
# proses RGB

import cv2 as cv
import numpy as np
import os
from imageio import imread
import pandas as pd


root_folder = 'F:/applications/python/Klasifikasi_Daun_Sawi/Dataset/testing'
path01 = 'F:/applications/python/Klasifikasi_Daun_Sawi/Dataset/testing/ADA HAMA'
path02 = 'F:/applications/python/Klasifikasi_Daun_Sawi/Dataset/testing/TANPA HAMA'
folders = [os.path.join(root_folder, x) for x in (path01, path01)]
all_images = [img for folder in folders for img in (folder)]
simpancsv = []
warna =[]
R = []
G = []
B = []

for cur_path in os.listdir(root_folder):
    src_sub_path = os.path.join(root_folder, cur_path)
    jdx = 0
    for filename in os.listdir(src_sub_path):
        filepath = os.path.join(src_sub_path, filename)
        image = cv.imread(filepath, cv.COLOR_RGB2BGR)
        rows, cols, _ = image.shape

        color_B = 0
        color_G = 0
        color_R = 0
        color_N = 0 # neutral/gray color
        for i in range(rows):
            
            for j in range(cols):
                k = image[i,j]
                if k[0] > k[1] and k[0] > k[2]:
                    color_B = color_B + 1
                    # warna.append(color_B)
                    continue
                if k[1] > k[0] and k[1] > k[2]:
                    color_G = color_G + 1
                    # warna.append(color_G)
                    continue
                if k[2] > k[0] and k[2] > k[1]:
                    color_R = color_R + 1
                    # warna.append(color_R)
                    continue
                color_N = color_N + 1
              
            pix_total = rows * cols
        G.append(color_G/pix_total)
        B.append(color_B/pix_total)
        R.append(color_R/pix_total)
        print("    ")
        print('Citra', filename,'Blue:', color_B/pix_total, 'Green:', color_G / pix_total,'Red:', color_R / pix_total)
        simpancsv.append(filename)
        # simpancsv.append(blue)
        
        
  
```
